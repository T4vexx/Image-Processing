# Noises models for images

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

img1 = cv2.imread("./images.jpeg", cv2.IMREAD_GRAYSCALE).astype(np.uint8)

fig=plt.figure(dpi=300)

def uniform_noise(img, max, pct=0.8):
  noisy_img = img.copy()
  x, y = img.shape
  mask = np.random.rand(x, y) < pct
  noisy_img[mask] = np.random.randint(0,max,size=noisy_img[mask].shape)
  return noisy_img

def gaussian_noise(img, mean, std):
  gauss = np.random.normal(mean, std, noisy_img.shape)
  return img + gauss

def salt_and_pepper_noise(img, salt_prob, pepper_prob):
    noisy_img = img.copy()
    x, y = img.shape
    
    # Criar máscara para sal
    salt_mask = np.random.rand(x, y) < salt_prob
    noisy_img[salt_mask] = 255  # Sal (branco)
    
    # Criar máscara para pimenta
    pepper_mask = np.random.rand(x, y) < pepper_prob
    noisy_img[pepper_mask] = 0  # Pimenta (preto)
    
    return noisy_img

noisy_img = uniform_noise(img1, 160)
gauss_img = gaussian_noise(img1, 20, 50)
salt_pepper_img = salt_and_pepper_noise(img1, 0.05, 0.05)

fig.add_subplot(1,4,1)
plt.imshow(img1,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,4,2)
plt.imshow(noisy_img,cmap='gray')
plt.axis("off")
plt.title("Uniforme")

fig.add_subplot(1,4,3)
plt.imshow(gauss_img,cmap='gray')
plt.axis("off")
plt.title("Gaussiano")

fig.add_subplot(1,4,4)
plt.imshow(salt_pepper_img,cmap='gray')
plt.axis("off")
plt.title("Salt and Pepper")

```
