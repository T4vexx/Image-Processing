```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

# Read the images
img1 = cv2.imread("./macaco.bmp", cv2.IMREAD_GRAYSCALE).astype(np.uint64)
img2 = cv2.imread("./lena.jpg", cv2.IMREAD_GRAYSCALE).astype(np.uint64)

# Initialize the result matrix
matResult = np.zeros_like(img1, dtype=np.uint64)
# Perform element-wise multiplication
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        matResult[i, j] = img1[i, j] * img2[i, j]

max = matResult.max()
min = matResult.min()

matResult = ((matResult - min) / (max - min) * 255).astype(np.uint8)

plt.imshow(matResult, cmap='gray')
plt.title("Image Multiplied")
plt.axis('off') 
plt.show()
```

```python
# Read the images
img1 = cv2.imread("./macaco.bmp", cv2.IMREAD_GRAYSCALE).astype(np.float32)
img2 = cv2.imread("./lena.jpg", cv2.IMREAD_GRAYSCALE).astype(np.float32)
# Initialize the result matrix
matResult = np.zeros_like(img1, dtype=np.float32)
# Perform element-wise multiplication
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
      if(img2[i, j] != 0):
        matResult[i, j] = img1[i, j] / img2[i, j]
      else:
        matResult[i, j] = img1[i, j]

max = matResult.max()
min = matResult.min()
matResult = ((matResult - min) / (max - min) * 255).astype(np.uint8)

plt.imshow(matResult, cmap='gray')
plt.title("Image Divided")
plt.axis('off')  
plt.show()
```
